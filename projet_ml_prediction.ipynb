```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Création d'un petit dataset fictif de maisons
# 'surface' en m², 'prix' en milliers d'euros
data = pd.DataFrame({
    'surface': [50, 70, 80, 100, 120],
    'prix': [100, 140, 160, 200, 240]
})

print("Aperçu du dataset :")
print(data.head())
```

```python
# Variable d'entrée X et variable cible y
X = data['surface'].values
y = data['prix'].values
```

```python
# Calcul du coefficient (a) et de l'intercept (b) de la régression linéaire
# Formule de base : a = covariance(X, y) / variance(X)
a = np.cov(X, y)[0,1] / np.var(X)
b = np.mean(y) - a * np.mean(X)

print(f"Modèle calculé : prix = {a:.2f}*surface + {b:.2f}")
```

```python
# Affichage graphique du dataset et de la régression
plt.scatter(X, y, color='blue', label="Données réelles")  # points réels
plt.plot(X, a*X + b, color='red', label="Régression prédite")  # ligne prédite
plt.xlabel("Surface (m²)")
plt.ylabel("Prix (k€)")
plt.title("Régression linéaire : surface → prix")
plt.legend()
plt.show()
```

```python
# Exemple de prédiction pour une maison de 90 m²
surface_test = 90
prix_pred = a*surface_test + b

print(f"Prix prédit pour {surface_test} m² : {prix_pred:.2f} k€")
```

```python
print("Conclusion :")
print("Le modèle permet de prédire le prix d'une maison selon sa surface.")
```
